```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)

plt.plot(x, np.sin(x),'-')
plt.plot(x, np.cos(x),'--')


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,100)

fig = plt.figure()

plt.scatter(x, np.sin(x))
plt.scatter(x, np.cos(x))


plt.show()

fig.savefig("my_figure.png")
```
