```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,2*np.pi,10)
fig=plt.figure()
plt.scatter(x,np.sin(x))
plt.scatter(x,np.cos(x));
plt.show()
```

```python
fig.savefig('Bushra')
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x));
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
plt.plot(x,np.tan(x));
```
