```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
fig.savefig('my_fig002.jpg')
```

```python
fig = plt.figure()
t=np.linspace(0,2*np.pi,100)

plt.subplot(2,2,1)
plt.plot(t,np.sin(t))

plt.subplot(2,2,2)
plt.plot(t,np.cos(t))

plt.subplot(2,2,3)
plt.plot(t,np.tan(t))

plt.subplot(2,2,4)
plt.plot(t,np.arctan(t))
fig.savefig("my_fig003.jpg")
```

```python
fig = plt.figure()
t=np.linspace(0,2*np.pi,10)

plt.plot(t,np.sin(t),'--ro',label='sine_func')
plt.plot(t,np.cos(t), '--b', label='cos_func')

#plt.xlim(-1,7)
#plt.ylim(-2,2)
#plt.axis([-1.7, -2.2])
#plt.axis('tight')

plt.grid()
plt.legend()
plt.title('This is heading', {'fontsize':20})
plt.xlabel('Horz axis', {'fontsize':15})
plt.ylabel('Vert axis', {'fontsize':15})
fig.savefig('my_fig3.jpg')
```

```python
x = np.linspace(0,10,50)
y = np.linspace(0,10,50)
fig = plt.figure(figsize=(10,7))

X,Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)

plt.contour(X,Y,Z,cmap='gist_rainbow')
# plt.contour(X,Y,Z,100,cmap='gist_rainbow')
# plt.controuf(X,Y,Z, cmap='gist_rainbow')

plt.colorbar()
fig.savefig('my_fig6.jpg')
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')

X,Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)

ax.contour3D(X,Y,Z,40, cmap = 'gist_rainbow')
fig.savefig('my_fig7.jpg')
```
