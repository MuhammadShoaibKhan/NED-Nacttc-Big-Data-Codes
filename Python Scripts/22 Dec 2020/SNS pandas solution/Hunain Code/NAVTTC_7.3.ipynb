```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,np.pi*2,100)
fig=plt.figure()
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
fig.savefig('my_fig0.jpg')
```

```python
fig=plt.figure()
t=np.linspace(0,2*np.pi,100)
plt.subplot(2,2,1)
plt.plot(t,np.sin(t))

plt.subplot(2,2,2)
plt.plot(t,np.cos(t))

plt.subplot(2,2,3)
plt.plot(t,np.tan(t))

plt.subplot(2,2,4)
plt.plot(t,np.arctan(t))
fig.savefig('my_fig1.jpg')
```

```python
fig=plt.figure()
t=np.linspace(0,2*np.pi,10)
plt.plot(t,np.sin(t),'--ro',label='sine_func')
plt.plot(t,np.cos(t),'--b*',label='cos_func')
#plt.xlim(-1,7)
#plt.ylim(-2,2)
#plt.axis([-1,7, -2,2])
#plt.axis('tight')
plt.grid()
plt.legend()
plt.title('THIS IS HEADING',{'fontsize':20})
plt.xlabel('Horz.axis',{'fontsize':15})
plt.ylabel('vert.axis',{'fontsize':15})
fig.savefig('my_fig2.jpg')
```

```python
x=np.linspace(0,10,50)
y=np.linspace(0,10,50)
fig=plt.figure(figsize=(10,8))
X,Y= np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
#plt.contour(X,Y,Z,cmap='gist_rainbow')
#plt.contour(X,Y,Z,cmap='gist_rainbow')
plt.contourf(X,Y,Z,cmap='bone')
plt.colorbar()
fig.savefig('my_fig3.jpg')
```

```python
from mpl_toolkits import mplot3d
fig=plt.figure(figsize=(10,7))
ax=plt.axes(projection='3d')
X,Y= np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
ax.contour3D(X,Y,Z,100,cmap='tab20c')
fig.savefig('my_fig4.jpg')
```

```python
#error plot

x=np.linspace(0,2*np.pi,25);
dy=0.25;

y= np.sin(x) + dy;
#y=np.sin(x) + dy*np.random.randn(25);
plt.errorbar(x,y,yerr=0.2, fmt='-o', color='g', ecolor='b', elinewidth=3, capsize=8);
```

```python
x=np.linspace(0,2*np.pi,25);
dx=0.15;
dy=0.25;

#y= np.sin(x) + dy;
y=np.sin(x) + dx*np.random.randn(25) + dy*np.random.randn(25);
plt.errorbar(x,y,xerr=dx, yerr=0, fmt='-o', color='b', ecolor='green', elinewidth=3, capsize=8);
plt.errorbar(x,y,xerr=0, yerr=dy, fmt='-o', color='b', ecolor='r', elinewidth=3, capsize=8);
```
