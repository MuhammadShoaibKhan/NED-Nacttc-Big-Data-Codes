```python
import seaborn as sns
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
rng = np.random.RandomState(0)
x = np.linspace(0,10,500)
y = np.cumsum(rng.randn(500,6),0)
```

```python
plt.plot(x,y)
plt.legend('ABCDEF', ncol=2, loc ='upper left');
```

```python
sns.set()
plt.plot(x,y)
plt.legend('ABCDEF', ncol=3, loc ='lower center');
```

```python
data = np.random.multivariate_normal([0,0],[[5,2],[2,2]], size =2000)
data = pd.DataFrame(data, columns=['x','y'])
for col in 'xy':
    #plt.hist(data[col],normed=True, alpha=0.5)
    plt.hist(data[col],density=True, alpha=0.5)
```

```python
plt.hist(data['x'],density=True, alpha=0.5, color ='grey');
plt.hist(data['y'],density=True, alpha=0.5, color ='purple');
plt.legend('xy', ncol=2, loc='upper left');
```

```python
for col in 'xy':
    sns.kdeplot(data[col], shade = True)
```

```python
for col in 'xy':
    sns.kdeplot(data[col], shade = False)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y']);
```

```python
sns.kdeplot(data, cmap ='viridis');
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',data, kind = 'kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',data, kind = 'scatter');
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',data, kind = 'reg');
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',data, kind = 'hex', color='g');
```
