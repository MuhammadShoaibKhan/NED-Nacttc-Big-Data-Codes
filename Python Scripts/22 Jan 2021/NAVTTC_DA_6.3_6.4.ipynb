```python
data['dur'] = data['dur'].str.strip('minutes')
data
```

```python
data['dur'].describe()
```

```python
data['dur'] = data['dur'].astype('int')
assert data['dur'].dtype == 'int'
```

```python
data['dur'].describe()
```

```python
data['veh_count'][data['veh_count'] > 25].size 
# checking how many values are greater than 25 in column 'veh_count'
```

```python
data['veh_count'][data['veh_count'] > 25].size
# again checking how many values are greater than 25 in column 'veh_count'
```

```python
import datetime as dt
today = dt.date.today()
today 
```

```python
data['travel_date'] = pd.to_datetime(data['travel_date']) 
```

```python
data['travel_date'].max()
```

```python
data.loc[data['travel_date'] > today, 'travel_date'] = today
# setting future dates to today's date in col 'travel_date'
```

```python
data['travel_date'].max()
```

```python
duplicates = data.duplicated(subset='travel_id', keep=False) # tells duplicate values in column 'travel_id'
duplicates # output is boolean (i.e. True or False)
```

```python
data['travel_id'][duplicates] # tells values which are duplicated in column 'travel_id'
```

```python
data['travel_id'][duplicates].size # tells counts of duplicated values in column 'travel_id'
```

```python
data['travel_id'][duplicates].sort_values()  # tells values which are duplicated (sorted) in column 'travel_id'
```

```python
duplicated_rides = data[duplicates].sort_values('travel_id')
duplicated_rides # duplicated rows w.r.t column 'travel_id'
```

```python
data['travel_id'][duplicates].sort_values().unique() 
# tells distinct values which are duplicated in column 'travel_id'
```

```python
data['travel_id'][duplicates].sort_values().unique().size
# tells counts of distinct values which are duplicated in column 'travel_id'
```

```python
data.shape # original data size before dropping duplicates rows
```

```python
# no. of original values = 25760

# no. of duplicated values = 16394 (in each: one unique and others are duplicates)

# no. of unique values in duplicated values = 6869 

# so finally total no. of distinct/unique values should be : 

# original(25760) - duplicates(16394) + unique_in_dupliates(6869)
# 25760 - 16394 + 6869
# 16235 
```

```python
data = data.drop_duplicates('travel_id') # dropping duplicate rows w.r.t column 'travel_id'  
data
```

```python
data.shape # data size after dropping duplicates rows w.r.t column 'travel_id' 
```

```python
import numpy as np
ranges = [0, 15, 30, np.inf]
tags = ['short', 'medium', 'long']
data['veh_count_categ'] = pd.cut(data['veh_count'] , bins = ranges, labels = tags) 
data
# values within range: 0 to 15 : assigns with tag name: 'short'
# values within range: 16 to 30 : assigns with tag name: 'medium'
# values from range: 31 and onwards : assigns with tag name: 'long'
```

```python
mappings = {'Male':'Men', 'Female':'Women'}
data['passen_boy_girl'] = data['passen_gender'].replace(mappings) 
data
# assigns 'Men' where value is 'Male'
# assigns 'Women' where value is 'Female'
```
