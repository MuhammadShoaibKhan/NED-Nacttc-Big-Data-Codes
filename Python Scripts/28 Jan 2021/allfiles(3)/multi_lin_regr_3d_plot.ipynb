```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv('multi_lin_reg_data_3d_plot.csv')
X = df[['Por', 'Brittle']].values
Y = df['Prod']

x = X[:, 0]  # x1
y = X[:, 1]  # x2
z = Y

pts = 20; 

x_pred = np.linspace(min(x), max(x), pts)   # range of x1 i.e. porosity values
y_pred = np.linspace(min(y), max(y), pts)   # range of x2 i.e. brittleness values
xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)

model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T

ols = linear_model.LinearRegression()
model = ols.fit(X, Y)
predicted = model.predict(model_viz)

r2 = model.score(X, Y)
coef = model.coef_
intr = model.intercept_

plt.style.use('default')

fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')

axes = [ax1, ax2, ax3]

for ax in axes:
    ax.plot(x, y, z, color='k', linestyle='none', marker='o', alpha=0.5)
    ax.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted, alpha=0.25)
    ax.set_xlabel('Porosity (%)', fontsize=12)
    ax.set_ylabel('Brittleness', fontsize=12)
    ax.set_zlabel('Gas Prod. (Mcf/day)', fontsize=12)

ax1.view_init(elev=30, azim=120)
ax2.view_init(elev=5, azim=100)
ax3.view_init(elev=60, azim=160)

print('Score:', np.round(r2*100,2), '%')
print('Coeff:', np.round(coef,2))
print('Intercept:', np.round(intr,2))
```

```python
from sklearn.linear_model import LinearRegression

df = pd.read_csv('multi_lin_data.csv')
x = df[['Por', 'Brittle']]
y = df['Prod']

# applying ML : 
reg = LinearRegression() # linear regression = formula : ypred = mx + b
reg.fit(x , y) # x must be a df  

m = reg.coef_; # m = slope
b = reg.intercept_; # b = y_intercept
acc = reg.score(x,y); # acc = accuracy 
ypred = reg.predict(x); # y_predicted

# displaying o/p : 
print('slope = m = ' , np.round(m,2)) # theta or coeff or slope
print('y_intercept = b = ' , np.round(b,2)) # y_intercept 
print('accuracy = ', np.round(acc*100 ,2) , '%') # accuracy
```
